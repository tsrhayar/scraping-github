```python
import csv
import requests
import sys
from dotenv import load_dotenv
import os

my_token = os.environ['TOKEN']

def save_repository_data_to_csv(query,num):
    url = 'https://api.github.com/search/repositories'

    headers = {
        'Authorization': f'Bearer {my_token}'
    }

    params = {
        'q': query,
        'sort': 'stars',
        'order': 'desc',
        'per_page': '100',
        'page': '1',
    }

    # Send the GET request
    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    total_count = data['total_count']

    # Save data in a CSV file
    filename = f'repositories2/repository_data{num}.csv'
    fieldnames = [
            'Name', 'Description', 'Stars', 'URL',
            'Date creation', 'Dernier modification',
            'Forks', 'Language'
            ]

    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        for item in data['items']:
            name = item['name']
            description = item['description']
            stars = item['stargazers_count']
            url = item['html_url']
            created_at = item['created_at']
            updated_at = item['updated_at']
            forks = item['forks_count']
            language = item['language']

            writer.writerow({
                'Name': name, 'Description': description,
                'Stars': stars, 'URL': url, 'Date creation': created_at,
                'Dernier modification': updated_at, 'Forks': forks, 'Language': language
                })

            
    print(f'CSV file "{filename}" created successfully!')
    return {"my_count":total_count, "less_star" : stars}
# ////////////////////////////
query = 'stars:>2500'
my_return = save_repository_data_to_csv(query,1)
my_count = my_return['my_count']
my_less_star =  my_return['less_star']
 print(f'Il reste {my_count}')

query = f'stars:2500..{my_less_star}'

my_counter = 2;
while my_count > 10:
    my_return = save_repository_data_to_csv(query,my_counter)
    my_count = my_return['my_count']
    my_less_star =  my_return['less_star']
    query = f'stars:2500..{my_less_star}'
    print(f'Il reste {my_count}')
    my_counter = my_counter + 1
    
# print('-----')

import glob
import pandas as pd

# Set the directory where the CSV files are located
directory = 'repositories2/'

# Set the pattern of the CSV files you want to merge
pattern = 'repository_data*.csv'

# Get the list of files in the directory that match the pattern
csv_files = glob.glob(directory + pattern)

# Create an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Iterate over the CSV files and merge them into the DataFrame
for file in csv_files:
    df = pd.read_csv(file)
    merged_data = pd.concat([merged_data, df], ignore_index=True)

# Save the merged data to a new CSV file
merged_data.to_csv('merged_file.csv', index=False)

print("CSV files merged successfully!")

# print('-----')
print('Terminé !')
```
